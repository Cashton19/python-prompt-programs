```python
# Clothes selector based on weather conditions
weather_condition = input("Enter the weather condition:(sunny, rainy, snowy, cloudy, windy)").strip().upper()
print(f"weather : {weather_condition}")
match weather_condition:
    case "RAINY":
        suggestion = "Take an umbrella and wear a raincoat!"
    case "SUNNY":
        suggestion = "Wear light clothing and sunglasses!"
    case "SNOWY":
        suggestion = "Wear warm clothes and boots!"
    case "CLOUDY":
        suggestion = "A light jacket might be useful"
    case "WINDY":
        suggestion = "Wear a windbreaker or jacket!"
    case _:
        suggestion = " Unknown weather condition."
print(f"Suggestion : {suggestion}")
    
```
