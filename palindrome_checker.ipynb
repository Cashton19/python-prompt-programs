```python
# Write a Python program to check if a number is a palindrome or no.  
try:
    s_number =input("Enter a number:")
    number = int(s_number)
    if s_number == s_number[::-1]:
        print(f"{number} is a palindrome")
    else:
        print(f"{number} is not a palindrome")
except ValueError:
    print("Error: invalid input")
```
