```python
def get_number(prompt):
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Error: Invalid Input, please enter a positive number.")
def get_maxnum(num1,num2,num3):
    max_num = max(num1,num2,num3)
    return max_num
def main():
    num1 = get_number("Enter number 1: ")
    num2 = get_number("Enter number 2: ")
    num3 = get_number("Enter number 3: ")
    max_num = get_maxnum(num1,num2,num3)
    print(f"Maximum of the the number {num1}, {num2} and {num3} is {max_num}")
if __name__ == "__main__":
    main()
```
