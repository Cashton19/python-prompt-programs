```python
import time,getpass,os
try:
    print("Writing to student.txt.....")
    time.sleep(2)
    whoami = getpass.getuser()
    os.makedirs(rf"C:\Users\{whoami}\Desktop\Output",exist_ok = True)
    with open(rf"C:\Users\{whoami}\Desktop\Output\student.txt","w") as file:
        file.write("Name: Cashton\n")
        file.write("Age: 19\n")
        file.write("Course: Computer Science")
    print("File Created Sucessfully.")
    with open(rf"C:\Users\{whoami}\Desktop\Output\student.txt","r") as file:
        content = file.read()
        print("Reading your file......")
        time.sleep(1)
        print(f"\n{content}")
        print("\nDone.")
except Exception as e:
    print(f"Error in file operations: {e}")
```
