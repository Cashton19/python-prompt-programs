```python
# lambda functions
add = lambda x,y: x + y 
subtract = lambda x, y: x - y
divide = lambda x, y: x / y
multiply = lambda x,y: x * y

def get_number(prompt):
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Error! Invalid input.")

def select_operator(num2):
    while True:
        operator = input("Please enter an operator for the calculation \n(+,-,*,/):").strip()
        if operator not in ["+","-","*","/"]:
            print("Error: Invalid input. \nSelect an appropriate operator.")
        elif operator == "/" and num2 == 0:
            print("Error: Denominator can't be zero.")
        else:
            return operator
        
def main():
    print("="*50)
    print("Arithmetic Calculator")
    print("="*50)
    num1 = get_number("Enter Number 1:")
    num2 = get_number("Enter Number 2:")
    operator = select_operator(num2)
    if operator == "+":
        result = add(num1,num2)
    elif operator == "-":
        result = subtract(num1,num2)
    elif operator == "*":
        result = multiply(num1, num2)
    elif operator == "/":
        result = divide(num1, num2)
    print(f"{num1} {operator} {num2} is {result:.2f}")
if __name__ == "__main__":
    main()
```
