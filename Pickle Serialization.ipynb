```python
import pickle, time, os, getpass, pprint
try:
    output_dir = rf"C:\Users\{getpass.getuser()}\Desktop\Output"
    os.makedirs(rf"C:\Users\{getpass.getuser()}\Desktop\Output", exist_ok = True)
    student_data = {
        "name": "Cashton Nadar",
        "age": 21,
        "Courses": ["Python", "Java", "Database"],
        "Grades": {"Python": "A",
                   "Java": "B+", 
                   "Database": "A-"}
    }
    print("Writing to student.pkl.....")
    time.sleep(2)
    with open(rf"{output_dir}\student.pkl", "wb") as file:
        pickle.dump(student_data, file)
    print("Write Sucessful.")
    time.sleep(2) 

    print("\nLoading student.pkl.....")
    with open(rf"{output_dir}\student.pkl", "rb") as file:
        loaded_data = pickle.load(file)
    print("Loaded data:")
    pprint.pprint(loaded_data)

except Exception as e:
    print(f"Error in file operations: {e}")
```
