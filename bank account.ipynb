```python
def create_bank_account(
    account_holder, 
    initial_balance = 0, 
    account_type = "savings", 
    interest_rate = 2.5, 
    minimum_balance = 100
):
    return{
        "account_holder" : account_holder,
        "initial_balance": initial_balance, 
        "account_type": account_type,
        "interest_rate": interest_rate, 
        "minimum_balance": minimum_balance
    }
def display_account(account):
    print(f"Account Holder: {account['account_holder']}")
    print(f"Intial Balance: \u20B9{account['initial_balance']:,.2f}")
    print(f"Account Type: {account['account_type']}")
    print(f"Interest Rate: {account['interest_rate']}\u0025")
    print(f"Minimum Balance: \u20B9{account['minimum_balance']:,.2f}")
def main():
    print("Account 1 (All defaults except name): ")
    account1 = create_bank_account("Rajesh Kumar")
    display_account(account1)
    print("-"*50)
    print("\nAccount 2 (All parameters specified):")
    account2 = create_bank_account("Priya Sharma",5000,"checking",1.5,50)
    display_account(account2)
    print("-"*50)
    print("\nAccount 3 (Mixed positional and keyword):")
    account3 = create_bank_account("Arjun Patel", 10000, account_type="Premium")
    display_account(account3)
    print("-"*50)
if __name__ == "__main__":
    main()
```
