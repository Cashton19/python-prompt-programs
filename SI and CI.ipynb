```python
def get_positive_number(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                print("Error: Please enter a Non-Zero or positive number.")
            else:
                return value
        except ValueError:
            print("Error: Invalid Input, please enter a positive number.")
def calculate_simple_interest(principal, time, rate_of_interest):
    SI = (principal * rate_of_interest * time)
    return SI
def calculate_compound_interest(principal, time, rate_of_interest):
    while True:
        compounding_frequency = get_positive_number(
            "Enter compounding frequency per year (1=annually, 4=quarterly, 12=monthly):"
        )
        if compounding_frequency not in [1, 4, 12]:
            print("Error: Enter only 1=annually, 4=quarterly, or 12=monthly.")
        else:
            break

    amount = principal * ((1 + (rate_of_interest / compounding_frequency)) ** (compounding_frequency * time))
    CI = amount - principal
    return amount, CI
def main():
    principal = get_positive_number("Enter principal amount: ")
    time = get_positive_number("Enter time period (years): ")
    rate_of_interest = get_positive_number("Enter annual interest rate (%): ") / 100

    SI = calculate_simple_interest(principal, time, rate_of_interest)
    print(f"Simple Interest: {SI:.2f}")
    print(f"Total Amount (SI): {(SI + principal):.2f}")
    
    print("\nFor Compound Interest") 
    amount, CI = calculate_compound_interest(principal, time, rate_of_interest)
    print(f"Compound Interest: {CI:.2f}") 
    print(f"Total Amount (CI): {amount:.2f}")
    print(f"Difference (CI - SI): {(CI - SI):.2f}")
if __name__ == "__main__":
    main()
```
