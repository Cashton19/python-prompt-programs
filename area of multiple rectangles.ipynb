```python
def get_positive_number(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                print("Error: Only enter positive numbers.")
            else:
                return value
        except ValueError:
            print("Error: Invaild input! Only enter positive numbers.")
```

```python
def get_positive_int(prompt):
    while True:
        try:
            value = int(input(prompt))
            if value <= 0:
                print("Error: Only enter Integers.")
            else:
                return value
        except ValueError:
            print("Error: Invaild input! Only enter Integers.")
```

```python
def calculate_rect_area(width, length):
    area = width * length 
    return area
```

```python
def main():
    rectangles = get_positive_int("How many rectangles do you want to calculate?")
    for reactangle in range(0,rectangles):
        print(f"\nRectangle {reactangle + 1}")
        length = get_positive_number("Enter length:")
        width = get_positive_number("Enter width:")
        area = calculate_rect_area(width, length)
        print(f"Area: {area:.2f} sq.units")
        area *= 2
    print(f"\nTotal area of all rectangles: {area:.2f} sq.units")
main()
```
