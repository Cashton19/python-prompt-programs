```python
def get_number(prompt):
    while True:
        try:
            value = int(input(prompt))
            value = abs(value)
            return value
        except ValueError:
            print("Error: Invalid Input, please enter a positive number.")
def get_factors(num):
    factors = []
    for numbers in range(num,0,-1):
        if num % numbers == 0:
            factors.append(numbers)
    return factors
def display_factors(factors, num):
    print(f"The factors of {num} are ", end="")
    if len(factors) > 1:
        print(", ".join(map(str, factors[:-1])) + f" and {factors[-1]}")
    else:
        print(factors[0])
def main():
    num = get_number("Enter your number:")
    factors = get_factors(num)
    display_factors(factors, num)
if __name__ == "__main__":
    main()
```
