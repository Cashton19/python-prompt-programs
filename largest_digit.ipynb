```python
# Write a Python program to find the largest digit in a number.  

try:
    num = int(input("Enter a number:"))
    abs_num = abs(num)
    s_num = str(num)
    a_num = list(s_num)
    largest_digit = max(a_num)
    print(f"The largest digit in {num} is: {largest_digit}")
except ValueError:
    print("Error: Invalid Input")
```
