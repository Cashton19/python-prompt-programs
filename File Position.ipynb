```python
import time,os,getpass
try:
    filepath = f"C:\\Users\\{getpass.getuser()}\\Desktop\\Output"
    os.makedirs(filepath, exist_ok = True)
    print("Writing student.txt.....")
    time.sleep(2)
    with open(f"{filepath}\\student.txt","w") as file:
        file.write("Name: Cashton\n")
        file.write("Age: 19\n")
        file.write("Grade: A\n")
        file.write("Course: Computer Science\n")
    print("File Created Sucessfully.")

    time.sleep(2)
    with open(f"{filepath}\\student.txt", "r+") as file:
        pos = file.tell()
        print(f"\nInitial Position: {pos}")
        print("\nReading Some Data.....")
        time.sleep(2)
        print(f"{file.read(10)}")
        pos = file.tell()
        print(f"\nPosition after reading 10 characters: {pos}")
        print("\nSeeking to Position 5 and Reading.....")
        time.sleep(2)
        file.seek(5)
        remaining_content = file.read()
        print(f"Reading from position 5: {remaining_content.strip()}")
        print("\nSeeking to Beginning and Reading.....")
        time.sleep(2)
        file.seek(0)
        pos = file.tell()
        print(f"{file.read()}")
        print(f"File position after seek(0): {pos}")
        time.sleep(2)
        print("\n"+"."*10 + "And that how you play with position of a file" + "."*10)
except Exception as e:
    print(f"Error in file operations: {e}")
```
