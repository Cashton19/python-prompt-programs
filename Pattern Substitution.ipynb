```python
import re
print("Pattern substituion:")
original = "Hello World! Hello Python"
print(f"\nOriginal String: {original}")
replaced = original.replace("Hello", "Hi")
print(f"Replaced String: {replaced}")
regex_replaced = re.sub(r"\b\w{5}\b", "[Redacted]", original)
print(f"Regex Replaced String: {regex_replaced}")
print(f"\nThe Regex Substituiton was: \"\\b\\w{{5}}\\b\"")
```
