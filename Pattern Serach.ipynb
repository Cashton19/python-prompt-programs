```python
import re
print("Basic Pattern Seacrhing")
text =  "The quick brown fox jumps over the lazy dog"
print(f"Text: \"{text}\"")

pattern1 = "fox"
pos = text.find(pattern1)
print(f"\nPattern '{pattern1}': {'Found at position ' + str(pos) if pos != -1 else 'Not Found'}")

pattern2 = "cat"
pos = text.find(pattern2)
print(f"Pattern '{pattern2}': {'Found at position ' + str(pos) if pos != -1 else 'Not Found'}")

four_letter_words = re.findall(r"\b[A-Za-z]{4}\b",text)
print(f"\nThe Regex pattern \"\\b[A-Za-z]{{4}}\\b\" is used to find the four letter words: {four_letter_words}")
```
