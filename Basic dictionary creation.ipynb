```python
print("=" * 50)
print("Basic Dictionary Operations")
print("=" * 50)
print("\n1. Dictionary Creation Methods: ")
student = {
    "name":"Cashton",
    "age" : 19,
    "grade" : "A" 
}

my_dict_usingdict = dict(name="Cashton", age=19, grade = "A")
my_dict_tuples = dict([("name","cashton"),("age",19),("grade","A")])
print(f"""Literal syntax: {student}
dict() constructor:{my_dict_usingdict}
From tuples: {my_dict_tuples}
""")
print("\n2. Accessing Elements:")
print(f"Student name: {student['name']}")
print(f"Student age: {student.get('age')}")
print(f"Student email: {student.get('email', 'Not Provided')} ")

print("\n3.Adding Elements:")
student["email"] = "cashtonnadar11@gmail.com"
student.update(dict(semester=1,university="XYZ university"))
print(f"After adding elements: {student}")

print("\n4.Modifying Elements:")
student["age"] = 21
student["grade"] = "A+"
student["semester"] = 3
print(f"After modification: {student}")

print("\n5.Deleting Elements:")
del student["semester"]
removed = student.pop("university")
print(f"Removed: \"semester\" and \"{removed}\"")
print(f"After deletion: {student}")

print("\n6. Key Existence Check:")
print(f"'name' in student: {'name' in student}")
print(f"'phone' in student: {'phone' in student} ")

print("\n7. Dictionary Information:")
print(f"Length: {len(student)}")
print(f"All Keys: {list(student.keys())}")
print(f"All Values: {list(student.values())}")
print(f"Both Keys and Values: {list(student.items())}")
```
