```python
import math, pprint
print("1.Even Squares of Numbers(1-10):")
even_squares = {x:x**2 for x in range(1,11) if x % 2 == 0}
print(even_squares)

print("\n2.Word lengths")
fruits = ["apple", "banana", "cherry", "date", "blueberry"]
long_fruits = {fruit: len(fruit) for fruit in fruits if len(fruit) > 5}
print(f"Fruits having more than 5 characters: {long_fruits}")

print("\n3.Scores of Students")
students =  ["Priya", "Amit", "Neha", "Vikram", "Ravi"]
scores =  [85, 92, 78, 96, 88]
high_scorers = {name: score for name, score in zip(students,scores) if score >= 90} 
print(f"Students with scores >= 90: {high_scorers}")
result = {name: ("Pass" if score >= 85 else "Fail" ) for name, score in zip(students,scores) }
print(f"Results: {result}")

print("\n4. Math Operations:")
math_operations = {x : {
    "square": x**2,
    "cube" : x ** 3,
    "sqrt" : round(math.sqrt(x),2),
    "factorial" : math.factorial(x) if x <= 8 else "Too Large", 
} for x in range(1,11)
}
pprint.pprint(math_operations)
```
