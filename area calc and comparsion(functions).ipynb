```python
def get_positive_number(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                print("Error: Please enter a Non-Zero or positive number.")
            else:
                return value
        except ValueError:
            print("Error: Invalid Input, please enter a positive number.")
def calculate_rect_area(length,width):
    rect_area = length * width
    return rect_area
def calculate_square_arae(side):
    sq_area = side * side
    return sq_area
def calculate_triangle_arae(base,height):
    tri_area = 0.5 * base * height
    return tri_area
def get_comparison(sq_area,rect_area,tri_area):
    dict_of_shape = {"Square": sq_area,"Rectangle": rect_area,"Triangle": tri_area}
    sorted_shapes = sorted(dict_of_shape.items(), key=lambda x: x[1], reverse=True)
    print(f"\nLargest area: {sorted_shapes[0][0]} ({sorted_shapes[0][1]} sq. units)")
    print(f"Areas in descending order: {sorted_shapes[0][0]} > {sorted_shapes[1][0]} > {sorted_shapes[2][0]}")
def main():
    print("="*50)
    print("Shape Area Calculator")
    print("="*50)
    # setting variables
    length = get_positive_number("Enter rectangle length:")
    width = get_positive_number("Enter rectangle breadth:")
    side = get_positive_number("Enter square side:")
    base = get_positive_number("Enter triangle base:")
    height = get_positive_number("Enter triangle height:")
    # calculations
    sq_area = calculate_square_arae(side)
    rect_area = calculate_rect_area(length,width)
    tri_area = calculate_triangle_arae(base,height)
    print("\nAreas calculated:")
    # display comparison
    print(f"Rectangle: {rect_area} units")
    print(f"Square: {sq_area} units")
    print(f"Triangle: {tri_area} units")
    get_comparison(sq_area,rect_area,tri_area)
main()
```
