```python
def get_int_number(prompt):
    while True:
        try:
            value = int(input(prompt))
            return value
        except ValueError:
            print("Error: Invalid Input, please enter a valid number.")
            
def get_float_number(prompt):
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Error: Invalid Input, please enter a valid number.")
            
def get_numlist(user_num):
    if user_num > 10:
        print("Opps, Only upto 10 numbers are available")
        return []  
    elif user_num <= 0 :
        print("You must enter a positive number.")
        return [] 
    elif user_num == 1 :
        print("You must enter a number greater than 1.")
        return [] 
    else:    
        num_list = []
        for x in range(int(user_num)):
            temp_num = get_float_number(f"Enter number {x+1}: ")
            num_list.append(temp_num)
        return num_list
    
def get_maxnum(num_list):
    max_num = max(num_list)
    return max_num
    
def main():
    user_num = get_int_number("How many numbers you want?: ")
    num_list = get_numlist(user_num)
    if num_list != []:
        max_num = get_maxnum(num_list)
        print(f"Maximum of {', '.join(map(str, num_list[:-1]))} and {num_list[-1]} is {max_num}")
if __name__ == "__main__":
    main()
```
