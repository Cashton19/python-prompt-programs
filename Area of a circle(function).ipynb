```python
import math
def get_positive_number(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                print("Error: Please enter a Non-Zero or positive number.")
            else:
                return value
        except ValueError:
            print("Error: Invalid Input, please enter a positive number.")
def calculate_area(radius):
    area = math.pi * (radius ** 2)
    return area
def calculate_circumference(radius):
    circumference = 2*math.pi*radius
    return circumference
def main():
    print("="*50)
    print("Radius and Circumference calculator")
    print("="*50)
    radius = get_positive_number("\nEnter the radius of the circle:")
    print(f"Circle with radius {radius:.2f}:")
    area = calculate_area(radius)
    circumference = calculate_circumference(radius)
    print(f"Area: {area:.2f} sq. units")
    print(f"Circumference: {circumference:.2f} units")
main()
```
