```python
import math,datetime,os,time,getpass,sys
def get_positive_angle(prompt):
    while True:
        try:
            value = int(input(prompt))
            if value > 360:
                print("Error: Angle can't be greater than 360°.")
            elif value <= 0:
                print("Error: Angle can't be 0 or less than 0°.")
            else:
                return value
        except ValueError:
            print("Error: Invalid Input, please enter a positive angle.")

def get_positive_int(prompt):
    while True:
        try:
            value = int(input(prompt))
            if value <= 0:
                print("Error: Please enter a Non-Zero or positive number.")
            else:
                return value
        except ValueError:
            print("Error: Invalid Input, please enter a positive number.")

def get_positive_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                print("Error: Please enter a Non-Zero or positive number.")
            else:
                return value
        except ValueError:
            print("Error: Invalid Input, please enter a positive number.")

def triangle_area_base_height(base, height):
    area = 0.5 * base * height
    return area

def triangle_area_three_sides(a, b, c):
    if a + b <= c or a + c <= b or b + c <= a:
        print("Invalid triangle! Sum of any two sides must be greater than the third side.")
        return None
    else:
        s = ((a + b + c)/2)
        area = math.sqrt(s*(s-a)*(s-b)*(s-c))
        return area

def triangle_area_two_sides_angle(a, b, angle):
    angle = math.radians(angle)
    area = 0.5 * a * b * (math.sin(angle))
    return area

def options_selection(option):
    while True:
        if option == 1:
            height = get_positive_float("Enter height:")
            base = get_positive_float("Enter base:")
            area = triangle_area_base_height(base, height) 
            opt_data = f" \nheight = {height} \nbase = {base}"
            return area, opt_data
        elif option == 2:
            a = get_positive_float("Enter side a:")
            b = get_positive_float("Enter side b:")
            c = get_positive_float("Enter side c:")
            area = triangle_area_three_sides(a, b, c) 
            opt_data = f" \nsidea = {a}\nsideb = {b}\nsidec = {c}"
            return area, opt_data
        elif option == 3:
            a = get_positive_float("Enter side a:")
            b = get_positive_float("Enter side b:")
            angle = get_positive_angle("Enter included angle in degrees°:")
            area = triangle_area_two_sides_angle(a, b, angle) 
            opt_data = f" \nsidea = {a} \nsideb = {b} \nangle = {angle}°"
            return area, opt_data
        else:
            print("Invalid choice!")
            area = None
            opt_data = None
            return area, opt_data
            
def write_file(data, filename = "output.txt"):
    whoami = getpass.getuser()
    now = datetime.datetime.now().strftime("\n"+("_"*120)+"\n%y-%m-%d(%H:%M:%S) \n")
    os.makedirs(rf"C:\Users\{whoami}\Desktop\output", exist_ok=True)
    print("creating file.....")
    time.sleep(2) 
    with open(rf"C:\Users\{whoami}\Desktop\output\{filename}", "a") as file:
        file.write(now + f"{'*'*100}"+f"\n{data}\n"+ f"{'*'*100}")
    print("Results have been logged")
    time.sleep(2) 
    print("Check output.txt on your desktop")
    
def main():
    print("="*50)
    print("Triangle Area Calculator")
    print("="*50)
    print("1. Base and Height")
    print("2. Three Sides (Heron's Formula)") 
    print("3. Two Sides and Included Angle")
    
    option = get_positive_int("\nEnter your options(1-3):")
    area, opt_data = options_selection(option)
    method = [(1,"Base and Height"),(2,"Three Sides (Heron's Formula)"),(3,"Two Sides and Included Angle")]
    
    if area is None:
        print("No area calculated due to invalid input.")
        data = "Invalid choice"
    else:
        print("calculating area....")
        time.sleep(1) 
        data = f"Area of triangle: {area:,.2f} sq. units \nMethod used: {method[option-1][0]} ({method[option-1][1]})" + f"\nused inputs" + opt_data
    write_file(data, filename = "output.txt")
    time.sleep(0.5)
    print("-"*25 + "BYE" + "-"*25)

if __name__ == "__main__":
    main()
```
