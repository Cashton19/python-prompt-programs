```python
def get_positive_number(prompt):
    while True:
        try:
            value = int(input(prompt))
            if value <= 0:
                print("Error: Only enter positive numbers.")
            else:
                return value
        except ValueError:
            print("Error: Invaild input! Only enter positive numbers.")
            
```

```python
def calculate_rect_area(width, length):
    area = width * length 
    return area
```

```python
def calculate_rect_perimeter(width, length):
    perimeter = 2 * (width + length)
    return perimeter
```

```python
def main():
    length = get_positive_number("Enter the length of the rectangle")
    width = get_positive_number("Enter the width of the rectangle")
    area = calculate_rect_area(width, length)
    perimeter = calculate_rect_perimeter(width, length)
    print(f"Rectangle with length {length} and width {width}:") 
    print(f"Area: {area} units")  
    print(f"Perimeter: {perimeter} units")  
main()
```
