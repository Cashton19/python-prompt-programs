```python
import time,getpass,os

def write_binary_data(numbers, filename,output_dir):
    """This function creates a file and writes data """
    with open(rf"{output_dir}\{filename}", "wb") as file:
        for number in numbers:
            file.write(number.to_bytes(4, byteorder = "big"))
    print("Writing data...")
    time.sleep(2)
    print("Data Written Sucessfully...")

def read_binary_data(filename,output_dir):
    """This function reads a file and displays data """
    with open(rf"{output_dir}\{filename}", "rb") as file:
        read_numbers = []
        while True:
            content = file.read(4)
            if not content:
                break
            read_numbers.append(int.from_bytes(content, byteorder = "big"))
    time.sleep(2)
    print("\nData Read Sucessfully...")
    print(f"Numbers : {read_numbers}")
    print("\nDone.")

def main():
    try:
        filename = "output.bin"
        numbers = [10, 20, 30, 40, 50]
        whoami = getpass.getuser()
        output_dir = rf"C:\Users\{whoami}\Desktop\Output"
        os.makedirs(output_dir, exist_ok=True)
        write_binary_data(numbers, filename, output_dir)
        read_binary_data(filename, output_dir)
    except Exception as e:
        print(f"Error in file operations: {e}")
if __name__ == "__main__":
    main()
```
